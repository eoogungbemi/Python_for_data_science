```python
#Scatter Plot
# create a figure and axis
fig, ax = plt.subplots()

# scatter the sepal_length against the sepal_width
ax.scatter(df['GRE'], df['Probability'])
# set a title and labels
ax.set_title('Admission Chance')
ax.set_xlabel('GRE')
ax.set_ylabel('Probability')
```

```python
# get columns to plot
columns = df.columns.drop(['Probability'])
# create x data
x_data = range(0, df.shape[0])
# create figure and axis
fig, ax = plt.subplots()
# plot each column
for column in columns:
    ax.plot(x_data, df[column], label=column)
# set title and legend
ax.set_title('Admission Dataset')
ax.legend()
```

```python
# Histogram
```

```python
# create figure and axis
fig, ax = plt.subplots()
# plot histogram
ax.hist(df['Probability'])
# set title and labels
ax.set_title('Chances of Admission')
ax.set_xlabel('Points')
ax.set_ylabel('Frequency')
```

```python
# Bar Chart 
```

```python
# create a figure and axis 
fig, ax = plt.subplots() 
# count the occurrence of each class 
data = df['Probability'].value_counts() 
# get x and y data 
points = data.index 
frequency = data.values 
# create bar chart 
ax.bar(points, frequency) 
# set title and labels 
ax.set_title('Chances of Admission') 
ax.set_xlabel('Points') 
ax.set_ylabel('Frequency')
```

```python
## Box plots
```

```python
# get correlation matrix
corr = df.corr()
fig, ax = plt.subplots()
# create heatmap
im = ax.imshow(corr.values)

# set labels
ax.set_xticks(np.arange(len(corr.columns)))
ax.set_yticks(np.arange(len(corr.columns)))
ax.set_xticklabels(corr.columns)
ax.set_yticklabels(corr.columns)

# Rotate the tick labels and set their alignment.
plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
         rotation_mode="anchor")

# Loop over data dimensions and create text annotations.
for i in range(len(corr.columns)):
    for j in range(len(corr.columns)):
        text = ax.text(j, i, np.around(corr.iloc[i, j], decimals=2),
                       ha="center", va="center", color="black") 
```
